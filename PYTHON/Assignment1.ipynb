<h3>Q3. Keywords check</h3>
<p>Import the keyword module and write a program that checks if a word is a Python keyword.</p>

```python
import keyword

word = input('Enter a word : ')
if keyword.iskeyword(word):
    print('Yes ', word, ' is a keyword')
else:
    print('Not a keyword')
```

<h3>Q8. Variables & Mutability </h3>
Explain mutable vs immutable with code showing list (mutable) and tuple (immutable)
behavior.

```python
# Mutable in python means we can change the value.
l = [1,'hello',2.25,True]
l[2] = 3
print(l)
```

```python
# Immutable in python means we cannot change the value.
t = (1,'hello',2.25,True)
t[2] = 4
print(t)
```

<h3>Q9. Input Validation (type casting + condition)</h3>
